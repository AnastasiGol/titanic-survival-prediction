## Dealing with missing values

```python
train_df.isnull().sum()
```

```python
def handle_unknown(df):
    new_df = df.copy()
    #dropping unknown data with high procent of missing values
    new_df= new_df.drop('Cabin', axis = 1,errors = 'ignore' )
    to_drop = new_df[new_df['Embarked'].isnull()].index
    new_df = new_df.drop(to_drop, axis  =0, errors = 'ignore')

    #dropping columns with unique data
    new_df = new_df.drop(['Ticket', 'Name', 'PassengerId'], axis = 1, errors = 'ignore')
    return new_df
    
```

### Percentage of missing values: 

```python
missing_percent = pd.DataFrame(np.round(train_df.isnull().sum().values / len(train_df), 2), index =train_df.isnull().sum().index, columns = ['%'] )
```

## Categorical

```python
for col in train_df.columns:
    print(f"n unique in {col}: {train_df[col].nunique()}")
```

```python
def handle_categorical(df):
    new_df = df.copy()
    new_df= pd.get_dummies(new_df, columns = ['Embarked', 'Sex'], drop_first = True)
    return new_df
    
```

## Age

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
def predict_age(df):
    #defining X_age, y_age
    age_to_pred = df[df['Age'].isnull()]
    age_train = df[~df['Age'].isnull()]
    X_age = age_train.drop(['Age', 'Survived'], axis = 1,  errors = 'ignore')
    y_age = age_train['Age']

    #split and train model, make predictions
    X_train, X_test, y_train, y_test = train_test_split(X_age, y_age, test_size=0.3)
    rfr = RandomForestRegressor()
    rfr.fit(X_train, y_train)

    X_to_pred = age_to_pred.drop(['Age', 'Survived'], axis=1, errors = 'ignore')
    pred_ages = rfr.predict(X_to_pred)

    inds = df[df['Age'].isnull()]['Age'].index
    return pred_ages 
    
```

## Final data transforming

```python
train_df = df.copy()
```

```python
def transform_data(df):
    df = handle_unknown(df)
    df = handle_categorical(df)
    

    pred_ages = predict_age(df)
    inds = df[df['Age'].isnull()]['Age'].index
    df.loc[inds, 'Age'] = pred_ages
    df['Family'] = df['SibSp'] + df['Parch']

    return df
```

```python
train_df = transform_data(train_df)
```

## Pipeline

Pipeline for data scaling and dealing with unknown data:

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
def make_pipeline(model):
    pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),   
    ("scaler", StandardScaler()),
    ("model", model)
])
    return pipe
```
